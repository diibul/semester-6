```python
import numpy as np
```

a. Create a 2D array based on the last 6 digits of your student ID (NIM). 

```python
# 202310370311268

matrix = np.array ([[3, 1, 1], [2, 6, 8]])

print("matrix:")
print(matrix)
```

b. Perform addition, subtraction, and multiplication operations on the matrix you created. 

```python
print("addition:")
print(matrix + matrix)

print("\nsubtraction:")
print(matrix - matrix)

print("\nmultiplication:")
print(matrix * 2)
```

c. Convert the matrix into a 1D shape. 

```python
matrix_1d = matrix.flatten()

print ("\nMatrix dalam bentuk 1D:")
print(matrix_1d)
```

d. Perform slicing on the matrix. 

```python
# Matrix 1D

print("\nAmbil baris pertama:")
print(matrix_1d[0])

print("\nAmbil baris keempat:")
print(matrix_1d[4])

print("\nAmbil 2 baris pertama:")
print(matrix_1d[0:2])

print("\nAmbil baris ke 1 sampai ke 4:")
print(matrix_1d[1:5])
```

```python
# Matrix 2D

print("Ambil baris pertama:")
print(matrix[0])

print("\nAmbil kolom kedua:")
print(matrix[:, 1])

print("\nAmbil 2 kolom pertama:")
print(matrix[:, :2])
```
