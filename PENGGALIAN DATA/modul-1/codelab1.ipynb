d. Remove float data types from the list.

```python
data_no_float = []

for item in data_list:
    filtered_tuple = tuple(x for x in item if not isinstance(x, float))
    data_no_float.append(filtered_tuple)

print("\nData setelah menghapus float:")
print(data_no_float)
```

e. Sort the elements in the list in descending order. 

```python
data_no_float.sort (reverse=True)

print("\nData setelah diurutkan secara descending:")
print(data_no_float)
```
