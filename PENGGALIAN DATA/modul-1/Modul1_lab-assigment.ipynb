**ASSIGNMENT 2**

Create a tuple containing a dictionary with 10 key-value pairs of courses and grades.
Among these 10 key-value pairs, ensure that at least 3 have empty values. Perform the following
operations:  
a. Remove incomplete entries or those with empty values.  

```python
courses = ({
    "Data Mining": 85,
    "Machine Learning": 90,
    "Artificial Intelligence": "",
    "Database": 88,
    "Statistics": "",
    "Data Science": 92,
    "Algorithms": 80,
    "Programming": "",
    "Computer Networks": 87,
    "Operating Systems": 89
},)

print(courses)
```

```python
data = courses[0]

cleaned = {k:v for k,v in data.items() if v != ""}

print("Data setelah menghapus nilai kosong:")
print(cleaned)
```

b. Count the number of entries remaining after deletion.  

```python
print("Jumlah data tersisa:", len(cleaned))
```

c. Sort the key-value pairs by value in ascending order.

```python
sorted_courses = dict(sorted(cleaned.items(), key=lambda x: x[1]))

print("Data setelah diurutkan:")
print(sorted_courses)
```

d. Find the course with the highest grade.  

```python
highest = max(cleaned, key=cleaned.get)

print("Mata kuliah dengan nilai tertinggi:", highest)
print("Nilainya:", cleaned[highest])
```

e. Calculate the average of all course grades.

```python
average = sum(cleaned.values()) / len(cleaned)

print("Rata-rata nilai:", average)
```

Create a dataframe consisting of 10 rows and 6 columns, with the following information: Screenshot 2026-03-07 094438.png

Perform the following analysis on the DataFrame:  
a. Display a summary of descriptive statistics including minimum, maximum, sum, mean,
standard deviation, and quartile 1, 2, and 3 values.

```python
import pandas as pd

data = {
    "Brand": ["Avoskin","G2G","Skintific","Wardah","Emina","Somethinc","Azarine","Lacoco","Y.O.U","Madame Gie"],
    "Branch": ["Surabaya","Jakarta Pusat","Bandung","Bandung","Jakarta","Bandung","Surabaya","Bandung","Jakarta Pusat","Surabaya"],
    "Price": [150,75,120,85,75,120,50,90,55,50],
    "Stock": [175,160,120,300,230,200,180,100,180,200],
    "Status": ["Sold","Available","Available","Sold","Available","Available","Available","Sold","Available","Sold"]
}

df = pd.DataFrame(data)

df
```

```python
df.describe()
```

```python
df[["Price","Stock"]].agg(["min","max","sum","mean","std"])
```

b. Identify the top 3 brands with the largest revenue (price x stock).

```python
df["Revenue"] = df["Price"] * df["Stock"]

top3 = df.sort_values(by="Revenue", ascending=False).head(3)

top3[["Brand","Revenue"]]
```

c. Group brands with the same price.

```python
df.groupby("Price")["Brand"].apply(list)
```

d. Create a new column containing categories 'Cheap', 'Medium', and 'Expensive'. These
categories are based on the following rules:  
- Prices below quartile 1 are categorized as 'Cheap'.  
- Prices above quartile 3 are categorized as 'Expensive'.  
- The rest are categorized as 'Medium'.

```python
Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)

def category(price):
    if price < Q1:
        return "Cheap"
    elif price > Q3:
        return "Expensive"
    else:
        return "Medium"

df["Category"] = df["Price"].apply(category)

df
```
