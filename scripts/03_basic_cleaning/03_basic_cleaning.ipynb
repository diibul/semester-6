Membersihkan teks dari noise dasar:
- Case folding (lowercase)
- Hapus URL, hashtag, mention
- Hapus karakter khusus dan angka
- Normalisasi huruf berulang
- Filter teks terlalu pendek (<10 karakter)
- Hapus duplikat

**Input:** `data/raw/ulasan_parfum_raw.csv`  
**Output:** `data/interim/ulasan_parfum_interim.csv`

```python
import sys
from pathlib import Path

ROOT_DIR = Path().resolve().parent.parent
sys.path.insert(0, str(ROOT_DIR))
print(f'Root: {ROOT_DIR}')
```

```python
# Demonstrasi fungsi-fungsi cleaning secara individual
import re

def case_folding(teks):
    """Ubah semua huruf menjadi huruf kecil."""
    return str(teks).lower().strip()

def hapus_url(teks):
    """Hapus URL (http, https, www)."""
    return re.sub(r'https?://\S+|www\.\S+', ' ', teks)

def hapus_karakter_khusus(teks):
    """Hapus tanda baca dan simbol non-alfanumerik."""
    return re.sub(r'[^\w\s]', ' ', re.sub(r'_', ' ', teks))

def hapus_angka(teks):
    """Hapus angka dalam teks."""
    return re.sub(r'\d+', ' ', teks)

def hapus_huruf_berulang(teks):
    """Normalisasi huruf yang diulang >2x. Contoh: 'waaangi' -> 'waangi'"""
    return re.sub(r'(.)\1{2,}', r'\1\1', teks)

def normalisasi_spasi(teks):
    return re.sub(r'\s+', ' ', teks).strip()

def basic_clean(teks):
    teks = case_folding(teks)
    teks = hapus_url(teks)
    teks = hapus_karakter_khusus(teks)
    teks = hapus_angka(teks)
    teks = hapus_huruf_berulang(teks)
    teks = normalisasi_spasi(teks)
    return teks

# Contoh transformasi
contoh = [
    'Waaanginya bgt!! check @parfumlokal https://toko.com',
    'Saff&co parfum ter the best 100% recommended!!!',
    'Ok bgt sih 5/5 stars'
]
print('Contoh Transformasi Basic Cleaning:')
print('-' * 60)
for c in contoh:
    print(f'  Sebelum: {c}')
    print(f'  Sesudah: {basic_clean(c)}')
    print()
```

```python
# Jalankan pipeline basic cleaning penuh
import importlib.util

script = ROOT_DIR / 'scripts' / '03_basic_cleaning' / 'basic_cleaning.py'
spec = importlib.util.spec_from_file_location('basic_cleaning', script)
mod  = importlib.util.module_from_spec(spec)
spec.loader.exec_module(mod)

df = mod.main()
```
