Membuktikan kualitas dataset yang dihasilkan dengan melatih dua model klasifikasi sentimen.

**Algoritma:**
- **SVM (LinearSVC)** dengan `class_weight='balanced'`
- **Multinomial Naive Bayes** dengan SMOTE pada data latih

**Representasi Fitur:** TF-IDF (unigram + bigram, max 5000 fitur)

**Input:** `data/processed/ulasan_parfum_processed.csv`  
**Output:** `results/models/`, `results/figures/`, `results/reports/ml_evaluation_report.txt`

```python
import sys
from pathlib import Path

ROOT_DIR = Path().resolve().parent.parent
sys.path.insert(0, str(ROOT_DIR))
print(f'Root: {ROOT_DIR}')
```

```python
# Jalankan seluruh pipeline machine learning
import importlib.util

script = ROOT_DIR / 'scripts' / '07_machine_learning' / 'modelling.py'
spec = importlib.util.spec_from_file_location('modelling', script)
mod  = importlib.util.module_from_spec(spec)
spec.loader.exec_module(mod)

mod.main()
```

```python
# Tampilkan laporan evaluasi
report_path = ROOT_DIR / 'results' / 'reports' / 'ml_evaluation_report.txt'
with open(report_path, 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# Visualisasi perbandingan model
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
gambar = [
    ('perbandingan_model.png', 'Perbandingan Model'),
    ('confusion_matrix_naive_bayes.png', 'Confusion Matrix - Naive Bayes'),
    ('confusion_matrix_svm.png', 'Confusion Matrix - SVM'),
]
for ax, (fname, judul) in zip(axes, gambar):
    img_path = ROOT_DIR / 'results' / 'figures' / fname
    if img_path.exists():
        ax.imshow(mpimg.imread(str(img_path)))
        ax.set_title(judul, fontsize=11, fontweight='bold')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Demo prediksi menggunakan model terbaik
import joblib

model_path = ROOT_DIR / 'results' / 'models' / 'model_terbaik_naive_bayes.pkl'
tfidf_path = ROOT_DIR / 'results' / 'models' / 'tfidf_vectorizer.pkl'

model = joblib.load(model_path)
tfidf = joblib.load(tfidf_path)

teks_baru = [
    'great smell long lasting very worth it',
    'bad smell disappointed not worth the price',
    'ok saja biasa'
]

X_baru = tfidf.transform(teks_baru)
prediksi = model.predict(X_baru)

print('Demo Prediksi Model Naive Bayes:')
print('-' * 60)
for teks, label in zip(teks_baru, prediksi):
    print(f'  Teks  : {teks}')
    print(f'  Label : {label}\n')
```

## Hasil Evaluasi

| Model | Akurasi | F1-Macro | F1-Weighted |
|---|---|---|---|
| Multinomial Naive Bayes | **92,53%** | **0,7102** | **0,9227** |
| SVM (LinearSVC) | 89,66% | 0,6163 | 0,8776 |

Model terbaik: **Multinomial Naive Bayes**
