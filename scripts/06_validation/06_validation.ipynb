**Input:** `data/processed/ulasan_parfum_processed.csv`

```python
import sys
from pathlib import Path

ROOT_DIR = Path().resolve().parent.parent
sys.path.insert(0, str(ROOT_DIR))
print(f'Root: {ROOT_DIR}')
```

```python
# Jalankan pipeline validasi
import importlib.util

script = ROOT_DIR / 'scripts' / '06_validation' / 'validation.py'
spec = importlib.util.spec_from_file_location('validation', script)
mod  = importlib.util.module_from_spec(spec)
spec.loader.exec_module(mod)

mod.main()
```

```python
# Tampilkan laporan validasi
report_path = ROOT_DIR / 'results' / 'reports' / 'validation_report.txt'
with open(report_path, 'r', encoding='utf-8') as f:
    print(f.read())
```
